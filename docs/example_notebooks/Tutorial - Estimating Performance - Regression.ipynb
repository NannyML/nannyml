```python
import nannyml as nml
from IPython.display import display

reference_df = nml.load_synthetic_car_price_dataset()[0]
analysis_df = nml.load_synthetic_car_price_dataset()[1]
display(reference_df.head(3))
```

```python
print(reference_df.head(3).to_markdown(tablefmt="grid"))
```

```python
estimator = nml.DLE(
    feature_column_names=['car_age', 'km_driven', 'price_new', 'accident_count', 'door_count', 'fuel', 'transmission'],
    y_pred='y_pred',
    y_true='y_true',
    timestamp_column_name='timestamp',
    metrics=['rmse', 'rmsle'],
    chunk_size=6000,
    tune_hyperparameters=False
)
```

```python
estimator.fit(reference_df)
results = estimator.estimate(analysis_df)
display(results.filter(period='analysis').to_df())
```

```python
print(results.filter(period='analysis').to_df().to_markdown(tablefmt="grid"))
```

```python
display(results.filter(period='reference').to_df())
```

```python
print(results.filter(period='reference').to_df().to_markdown(tablefmt="grid"))
```

```python
for metric in estimator.metrics:
    metric_fig = results.plot(kind='performance', metric=metric, plot_reference=True)
    metric_fig.show()
```
