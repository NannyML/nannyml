```python
import nannyml as nml
from IPython.display import display

reference_df, monitored_df, monitored_targets_df = nml.load_synthetic_car_loan_dataset()
display(reference_df.head())
```

```python
from docs.utils import print_multi_index_markdown
print_multi_index_markdown(reference_df.head())
```

```python
calc = nml.SummaryStatsRowCountCalculator(timestamp_column_name='timestamp', chunk_period='M')
```

```python
calc.fit(reference_df)
results = calc.calculate(monitored_df)
display(results.filter(period='all').to_df())
```

```python
print_multi_index_markdown(results.filter(period='all').to_df())
```

```python
results.filter(period='monitored').plot().show()
```
